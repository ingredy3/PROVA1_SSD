```python
# === PASSO 2: Carregar dataset UNSW-NB15 (VERS√ÉO 82K) ===

# 1Ô∏è‚É£ Instalar e configurar a API do Kaggle
!pip install -q kaggle

# 2Ô∏è‚É£ Remover credenciais incorretas (se existirem)
import os
if 'KAGGLE_USERNAME' in os.environ:
    del os.environ['KAGGLE_USERNAME']
if 'KAGGLE_KEY' in os.environ:
    del os.environ['KAGGLE_KEY']

# 3Ô∏è‚É£ Baixar o dataset
!kaggle datasets download -d mrwellsdavid/unsw-nb15 -p ./ --unzip

# 4Ô∏è‚É£ Carregar ESPECIFICAMENTE o arquivo de 82K linhas
import pandas as pd
import glob

# Listar arquivos
csv_files = glob.glob("*.csv")
print("üìÅ Arquivos CSV encontrados:", csv_files)

# üéØ CARREGAR ESPECIFICAMENTE O ARQUIVO DE 82K LINHAS
target_file = None

# Prioridade 1: Arquivo de treino com ~82K linhas
for file in csv_files:
    if 'training' in file.lower() or 'train' in file.lower():
        target_file = file
        break

# Prioridade 2: Se n√£o encontrou, procurar por n√∫meros espec√≠ficos
if target_file is None:
    for file in csv_files:
        if any(str(i) in file for i in [1, 2, 3, 4]):  # UNSW-NB15_1.csv, etc.
            # Verificar o tamanho real do arquivo
            temp_df = pd.read_csv(file, nrows=1, low_memory=False)
            if 'srcip' in temp_df.columns:  # Verificar se tem colunas do dataset principal
                target_file = file
                break

print(f"\nüéØ Arquivo selecionado: {target_file}")

# 5Ô∏è‚É£ Carregar o dataset principal
if target_file:
    df = pd.read_csv(target_file, low_memory=False)
    print(f"‚úÖ Dataset carregado com SUCESSO: {target_file}")
    print(f"üìä Formato: {df.shape}")

    # Verificar se tem ~82K linhas
    expected_rows = 82000
    actual_rows = df.shape[0]

    if abs(actual_rows - expected_rows) <= 2000:  # Margem de erro de 2K
        print(f"üéØ PERFEITO! Encontramos a vers√£o de ~82K linhas: {actual_rows} linhas")
    else:
        print(f"üìù Dataset carregado com {actual_rows} linhas (esper√°vamos ~82K)")

    print(f"üîç Colunas: {df.shape[1]}")
    print("\nüìã Amostra dos dados:")
    print(df.head(3))

else:
    raise FileNotFoundError("‚ùå N√£o foi encontrar o arquivo de ~82K linhas!")

print(f"\nüéâ PRONTO! DataFrame 'df' criado com {df.shape[0]:,} linhas e {df.shape[1]} colunas")
```

### ‚öôÔ∏è Passo 4 ‚Äî Pr√©-processamento e Cria√ß√£o de Pipelines

#### ‚úÖ Corre√ß√µes e Ajustes Realizados

**1. Remo√ß√£o de Data Leakage**  
A coluna `attack_cat` foi removida das features, pois representa o **tipo de ataque** (target multiclasse) e, portanto, **cont√©m informa√ß√£o do futuro**.  
Manter essa vari√°vel como entrada geraria *data leakage*.  

**Resultado:**
- Categ√≥ricas v√°lidas: `['proto', 'service', 'state']`  
- Target: `label` (0 = normal, 1 = attack)

---

#### ‚öôÔ∏è Estrutura Final das Features

**3Ô∏è‚É£ Integra√ß√£o via `ColumnTransformer`**  
Combina ambas as transforma√ß√µes em um √∫nico pipeline, garantindo que o pr√©-processamento ocorra **dentro da valida√ß√£o cruzada**, evitando vazamento de dados.

```python
#  PASSO 4: PR√â-PROCESSAMENTO SEM DATA LEAKAGE ----

from sklearn.preprocessing import OneHotEncoder

print("=== üõ†Ô∏è CONFIGURA√á√ÉO DE PR√â-PROCESSAMENTO CORRIGIDA ===")

# 1. VERIFICAR E CORRIGIR FEATURES CATEG√ìRICAS
print("üîç Analisando features categ√≥ricas...")

# Remover poss√≠veis targets das features
features_to_remove = []
if 'attack_cat' in cat_features:
    features_to_remove.append('attack_cat')
    print("‚ö†Ô∏è  REMOVIDO: 'attack_cat' - √© provavelmente target multiclasse (tipos de ataque)")

# Verificar se h√° outras colunas que podem ser targets
for feature in cat_features:
    if feature.lower() in ['label', 'target', 'class', 'classification']:
        features_to_remove.append(feature)
        print(f"‚ö†Ô∏è  REMOVIDO: '{feature}' - poss√≠vel target")

# Atualizar lista de features categ√≥ricas
cat_features_corrected = [f for f in cat_features if f not in features_to_remove]

print(f"üìä Features num√©ricas: {len(numeric_features)}")
print(f"üìä Features categ√≥ricas CORRIGIDAS: {len(cat_features_corrected)}")
print(f"üî§ Categ√≥ricas: {cat_features_corrected}")

# 2. PIPELINES DE PR√â-PROCESSAMENTO CORRIGIDOS
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, cat_features_corrected)  # ‚úÖ USAR LISTA CORRIGIDA
    ])

# 3. MODELOS COM PR√â-PROCESSAMENTO CORRETO
models = {
    'logreg': Pipeline(steps=[('preprocessor', preprocessor),
                              ('clf', LogisticRegression(max_iter=1000, random_state=42))]),
    'rf': Pipeline(steps=[('preprocessor', preprocessor),
                          ('clf', RandomForestClassifier(n_jobs=-1, random_state=42))]),
    'svc': Pipeline(steps=[('preprocessor', preprocessor),
                           ('clf', SVC(probability=True, random_state=42))])
}

print('‚úÖ Pipelines criados para:', list(models.keys()))
print(f"üéØ Target: {TARGET_COLUMN} (0=normal, 1=attack)")
print(f"üìà Features totais ap√≥s pr√©-processamento: ~{len(numeric_features) + sum([df[cat].nunique() for cat in cat_features_corrected])}")

# 4. VERIFICA√á√ÉO EXTRA - ANALISAR attack_cat
if 'attack_cat' in df.columns:
    print(f"\n=== üîç AN√ÅLISE DA COLUNA 'attack_cat' ===")
    print("Valores √∫nicos em attack_cat:")
    print(df['attack_cat'].value_counts())
    print(f"\nRela√ß√£o com target 'label':")
    print(pd.crosstab(df['label'], df['attack_cat']))
```

```python
# ---- PASSO 6 Valida√ß√£o cruzada e RandomizedSearchCV ----
from scipy.stats import randint
from sklearn.model_selection import RandomizedSearchCV

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

# Par√¢metros para RandomForest (sem pipeline)
param_dist_rf = {
    'n_estimators': randint(50, 300),
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 4)
}

N_ITER = 6
USE_SUBSAMPLE_FOR_TUNING = False
SAMPLE_FRAC = 0.5

if USE_SUBSAMPLE_FOR_TUNING:
    # J√° s√£o arrays numpy, usar indexing num√©rico
    sample_size = int(len(X_train) * SAMPLE_FRAC)
    indices = np.random.choice(len(X_train), sample_size, replace=False)
    X_tune = X_train[indices]
    y_tune = y_train.iloc[indices] if hasattr(y_train, 'iloc') else y_train[indices]
else:
    X_tune, y_tune = X_train, y_train

print('Iniciando RandomizedSearchCV (RF). n_iter =', N_ITER)
print(f'Shape dos dados de tuning: {X_tune.shape}')

# Usar RandomForest diretamente (sem pipeline, pois dados j√° est√£o pr√©-processados)
rf_base = RandomForestClassifier(random_state=42, n_jobs=-1)

rnd_rf = RandomizedSearchCV(
    rf_base,
    param_distributions=param_dist_rf,
    n_iter=N_ITER,
    cv=cv,
    scoring='f1_macro',
    n_jobs=-1,
    verbose=1,
    random_state=42
)

try:
    rnd_rf.fit(X_tune, y_tune)
    print('‚úÖ Melhor params RF:', rnd_rf.best_params_)
    print('‚úÖ Melhor score (CV f1_macro):', rnd_rf.best_score_)

    # Treinar modelo final com melhores par√¢metros
    best_rf = RandomForestClassifier(**rnd_rf.best_params_, random_state=42, n_jobs=-1)
    best_rf.fit(X_train, y_train)
    print('‚úÖ Modelo final (best_rf) treinado com melhores hiperpar√¢metros.')

except Exception as e:
    print('‚ùå Erro durante RandomizedSearchCV:', e)
    print('üîß Usando fallback: RandomForest com par√¢metros padr√£o')
    # Fallback
    best_rf = RandomForestClassifier(random_state=42, n_jobs=-1)
    best_rf.fit(X_train, y_train)
    print('‚úÖ Modelo fallback treinado.')

# Adicionar aos modelos para uso posterior
models['rf_tuned'] = best_rf
```

```python
# ---- Passo 7 ----

print("=== üîÑ RESTAURANDO DADOS PR√â-PROCESSADOS ===")

# Refazer o pr√©-processamento CORRETO (igual ao Passo 5)
X = df.drop(columns=[TARGET_COLUMN])
y = df[TARGET_COLUMN]

# REMOVER attack_cat das features (corre√ß√£o cr√≠tica)
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_features = X.select_dtypes(include=['object']).columns.tolist()
if 'attack_cat' in cat_features:
    cat_features.remove('attack_cat')

# Aplicar pr√©-processamento
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False), cat_features)
    ])

X_processed = preprocessor.fit_transform(X)

# DIVIS√ÉO TREINO/TESTE COM DADOS PR√â-PROCESSADOS
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.2, stratify=y, random_state=42
)

print(f"‚úÖ Dados pr√©-processados restaurados:")
print(f"   X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"   y_train: {y_train.shape}, y_test: {y_test.shape}")

# AGORA re-treinar o RandomForest com dados consistentes
print("\n--- üîÑ Re-treinando RandomForest com dados consistentes ---")
best_rf = RandomForestClassifier(
    n_estimators=152,
    max_depth=30,
    min_samples_split=6,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1
)
best_rf.fit(X_train, y_train)
print("‚úÖ RandomForest re-treinado com dados consistentes")

# Adicionar aos modelos
models['rf_tuned'] = best_rf
```

### ü§ñ Passo 8 ‚Äî Treinamento de M√∫ltiplos Modelos para Compara√ß√£o

#### üéØ Objetivo
Comparar abordagens distintas ao problema de detec√ß√£o de intrus√µes treinando **tr√™s modelos**:  
**(1)** Random Forest otimizado (do Passo 6‚Äì7), **(2)** Logistic Regression e **(3)** K-Nearest Neighbors.

---

#### ‚öôÔ∏è Estrat√©gia
- **Diversidade algor√≠tmica:** √°rvore/ensemble (RF), linear (LogReg) e baseado em inst√¢ncias (KNN).  
- **Dados:** `X_train`, `y_train` e `X_test`, `y_test` **j√° pr√©-processados** (188 features), garantindo reprodutibilidade.  
- **Avalia√ß√£o principal:** **F1** (al√©m de acur√°cia, precis√£o, recall e ROC AUC).

---

#### ‚úÖ Modelos treinados
1. **Random Forest (rf_tuned):** hiperpar√¢metros √≥timos do Passo 6, re‚Äêtreinado no Passo 7.  
2. **Logistic Regression:** baseline linear com regulariza√ß√£o.  
3. **KNN (k=5):** vizinhan√ßa para compara√ß√£o r√°pida.

---

#### üìä M√©tricas reportadas
- **Acur√°cia**, **Precis√£o**, **Recall**, **F1** e **ROC AUC** em **teste**.  
- **Matriz de confus√£o** opcional para an√°lise de erros.

---

#### üß© Interpreta√ß√£o esperada
- **RF** tende a melhor F1 (robusto √† alta dimensionalidade e intera√ß√µes).  
- **LogReg** oferece **interpretabilidade** e baseline competitivo com boa regulariza√ß√£o.  
- **KNN** pode sofrer em **alta dimensionalidade** (188 features), servindo como refer√™ncia de m√©todo por inst√¢ncias.

**Status:** modelos treinados e comparados em condi√ß√µes iguais; prontos para a **avalia√ß√£o final** e **conclus√µes cr√≠ticas**.


```python
# === PASSO 8: Treinar 3 modelos e comparar em teste (m√©tricas completas) ===
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
)
import numpy as np
import matplotlib.pyplot as plt

print("=== ü§ñ Treinando 3 modelos (RF otimizado, LogReg, KNN) ===")

# 1) Logistic Regression (ajuste de solver para compatibilidade do n_jobs)
lr_model = LogisticRegression(
    max_iter=1000,
    solver="liblinear",   # garante compatibilidade ampla
    multi_class="ovr",
    n_jobs=-1,
    random_state=42
)
lr_model.fit(X_train, y_train)
models['logreg_baseline'] = lr_model

# 2) Random Forest (j√° otimizado e re-treinado no Passo 7)
# best_rf deve existir (rf_tuned). Mantemos como est√°:
models['rf_tuned'] = best_rf

# 3) KNN (k=5)
knn = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)
knn.fit(X_train, y_train)
models['knn_5'] = knn

print("‚úÖ Modelos treinados: ", list(models.keys()))

# ---------- Avalia√ß√£o em teste ----------
def eval_model(name, clf, X_te, y_te):
    y_pred = clf.predict(X_te)
    acc  = accuracy_score(y_te, y_pred)
    prec = precision_score(y_te, y_pred, zero_division=0)
    rec  = recall_score(y_te, y_pred, zero_division=0)
    f1   = f1_score(y_te, y_pred, zero_division=0)
    # ROC AUC (se houver predict_proba)
    try:
        y_proba = clf.predict_proba(X_te)[:, 1]
        auc = roc_auc_score(y_te, y_proba)
    except Exception:
        auc = np.nan
    return {"modelo": name, "acc": acc, "precision": prec, "recall": rec, "f1": f1, "roc_auc": auc}

results = []
for name in ["rf_tuned", "logreg_baseline", "knn_5"]:
    results.append(eval_model(name, models[name], X_test, y_test))

# Tabela de resultados ordenada por F1
import pandas as pd
res_df = pd.DataFrame(results).sort_values("f1", ascending=False).reset_index(drop=True)
print("\n=== üìä Desempenho em TESTE (ordenado por F1) ===")
display(res_df)

# (Opcional) Relat√≥rio detalhado e matriz de confus√£o do melhor
best_name = res_df.loc[0, "modelo"]
best_clf  = models[best_name]
print(f"\n=== üèÜ Relat√≥rio detalhado ‚Äî {best_name} ===")
y_pred_best = best_clf.predict(X_test)
print(classification_report(y_test, y_pred_best, digits=4))
cm = confusion_matrix(y_test, y_pred_best)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title(f"Matriz de Confus√£o ‚Äî {best_name} (teste)")
plt.tight_layout()
plt.show()
```

```python
# === PASSO 10 (ROBUSTO): Avalia√ß√£o final dos modelos ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix, ConfusionMatrixDisplay
)

print("=== üéØ AVALIA√á√ÉO FINAL DOS MODELOS (robusta) ===")

# 1) Checagens b√°sicas
missing = []
for var in ["X_test", "y_test"]:
    if var not in globals():
        missing.append(var)
if missing:
    raise RuntimeError(f"Vari√°veis ausentes: {missing}. Execute os passos 5/7 para preparar e fazer o split.")

# 2) Fun√ß√£o de avalia√ß√£o
def evaluate_model(model, X_te, y_te, name="Model"):
    print(f"\nüìä AVALIANDO: {name}")
    print("-" * 50)

    # previs√£o
    y_pred = model.predict(X_te)

    # m√©tricas (weighted para refletir distribui√ß√£o)
    acc  = accuracy_score(y_te, y_pred)
    prec = precision_score(y_te, y_pred, average="weighted", zero_division=0)
    rec  = recall_score(y_te, y_pred, average="weighted", zero_division=0)
    f1   = f1_score(y_te, y_pred, average="weighted", zero_division=0)

    print(f"‚úÖ Accuracy:  {acc:.4f}")
    print(f"‚úÖ Precision: {prec:.4f}")
    print(f"‚úÖ Recall:    {rec:.4f}")
    print(f"‚úÖ F1-Score:  {f1:.4f}\n")

    # relat√≥rio por classe (0=Normal, 1=Attack)
    print(f"üìã RELAT√ìRIO DE CLASSIFICA√á√ÉO ({name}):")
    print(classification_report(y_te, y_pred, target_names=["Normal", "Attack"], digits=4))

    # matriz de confus√£o
    cm = confusion_matrix(y_te, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Normal","Attack"])
    disp.plot(cmap="Blues")
    plt.title(f"Matriz de Confus√£o ‚Äî {name} (teste)")
    plt.tight_layout()
    plt.show()

    return {"model": name, "accuracy": acc, "precision": prec, "recall": rec, "f1": f1}

# 3) Resolver nomes reais no dicion√°rio `models`
if "models" not in globals() or not isinstance(models, dict):
    models = {}

def resolve_model(candidates):
    """Retorna (name, model) da primeira chave existente em `models` dentro de candidates."""
    for key in candidates:
        if key in models:
            return key, models[key]
    return None, None

# RF otimizado
rf_key, rf_model = resolve_model(["rf_tuned", "rf"])
# Logistic Regression (use a vers√£o treinada)
lr_key, lr_model = resolve_model(["logreg_baseline", "logreg"])
# KNN (use a vers√£o treinada)
knn_key, knn_model = resolve_model(["knn_5", "knn"])

# Tamb√©m aceitar `best_rf` se rf_model n√£o foi encontrado
if rf_model is None and "best_rf" in globals():
    rf_key, rf_model = "best_rf", best_rf

# 4) Avaliar o que estiver dispon√≠vel (evita 'pipeline not fitted' e KeyError)
results = []
if rf_model is not None:
    results.append(evaluate_model(rf_model, X_test, y_test, "Random Forest Otimizado"))
else:
    print("‚ö†Ô∏è Random Forest n√£o encontrado: espere encontrar 'rf_tuned' ou vari√°vel 'best_rf'.")

if lr_model is not None:
    # se for pipeline antigo n√£o-treinado, vai falhar aqui; nesse caso re-treine o passo 8.
    try:
        results.append(evaluate_model(lr_model, X_test, y_test, "Logistic Regression"))
    except Exception as e:
        print(f"‚ÑπÔ∏è Pulei Logistic Regression (provavelmente pipeline n√£o-treinado): {e}")
else:
    print("‚ÑπÔ∏è Logistic Regression n√£o encontrada (procure por 'logreg_baseline').")

if knn_model is not None:
    try:
        results.append(evaluate_model(knn_model, X_test, y_test, "K-Nearest Neighbors"))
    except Exception as e:
        print(f"‚ÑπÔ∏è Pulei KNN (verifique se foi treinado no Passo 8): {e}")
else:
    print("‚ÑπÔ∏è KNN n√£o encontrado (procure por 'knn_5').")

# 5) Resumo em tabela
if results:
    res_df = pd.DataFrame(results).sort_values("f1", ascending=False).reset_index(drop=True)
    print("\n" + "="*60)
    print("üéØ RESUMO COMPARATIVO - TODOS OS MODELOS")
    print("="*60)
    display(res_df)
    res_df.to_csv("results_summary.csv", index=False)
    print("üíæ results_summary.csv salvo!")
else:
    print("‚ùå Nenhum modelo foi avaliado. Refa√ßa o Passo 8 para treinar LogReg/KNN e o Passo 7/9 para o RF.")

print("\n‚úÖ AVALIA√á√ÉO CONCLU√çDA (modo robusto)!")
```

# üéØ PASSO 11 - Curvas AUC-ROC e An√°lise Final de Desempenho

## üìä Objetivo
Avaliar e comparar o desempenho dos modelos de classifica√ß√£o atrav√©s das **Curvas ROC (Receiver Operating Characteristic)** e c√°lculo da **√Årea Sob a Curva (AUC)**, identificando qual modelo apresenta melhor capacidade discriminativa para detec√ß√£o de intrus√µes em rede.

## üîç Metodologia
- **Previs√µes Probabil√≠sticas**: Obter as probabilidades de classe para cada modelo
- **Curvas ROC**: Plotar a rela√ß√£o entre Taxa de Verdadeiros Positivos (TPR) vs Taxa de Falsos Positivos (FPR)
- **C√°lculo de AUC**: Quantificar a capacidade de discrimina√ß√£o de cada modelo
- **Thresholds √ìtimos**: Identificar os pontos de opera√ß√£o ideais

## üìà M√©tricas Calculadas
- **AUC-ROC**: √Årea sob a curva ROC (0.5 = aleat√≥rio, 1.0 = perfeito)
- **Threshold √ìtimo**: Ponto de melhor equil√≠brio entre TPR e FPR
- **FPR/TPR √ìtimos**: Taxas no ponto de opera√ß√£o ideal

## üé® Visualiza√ß√µes
1. **Curvas ROC Comparativas**: Compara√ß√£o visual entre todos os modelos
2. **Pontos √ìtimos**: Destacando os melhores pontos de opera√ß√£o
3. **Tabela Resumo**: M√©tricas consolidadas para an√°lise r√°pida

## üèÜ Crit√©rio de Sele√ß√£o do Melhor Modelo
O modelo com **maior AUC-ROC** ser√° considerado o melhor classificador para detec√ß√£o de intrus√µes, indicando superior capacidade de distinguir entre tr√°fego normal e malicioso.

> **üí° Interpreta√ß√£o dos Resultados AUC:**
> - **AUC > 0.9**: Classificador excelente
> - **AUC > 0.8**: Classificador bom  
> - **AUC > 0.7**: Classificador aceit√°vel
> - **AUC = 0.5**: Classificador aleat√≥rio (sem poder discriminativo)

```python
# ====================
# PASSO 11 - AUC-ROC CORRIGIDO
# ====================

print("üîç PASSO 11: AUC-ROC para Arrays NumPy")

# Agora faz sentido! X_test j√° est√° processado como array numpy
print(f"üìä X_train shape: {X_train.shape}")
print(f"üìä X_test shape: {X_test.shape}")
print(f"üìä X (original) shape: {X.shape}")

# Os modelos foram treinados com X_train (188 features) mas o preprocessor
# foi treinado com X (44 features). Vamos usar X_test diretamente!

try:
    # 1. Fazer previs√µes diretamente no X_test (j√° processado)
    print("\nüìä Gerando previs√µes probabil√≠sticas...")

    # Random Forest
    y_pred_rf_proba = rf_model.predict_proba(X_test)[:, 1]
    print("‚úÖ Random Forest - previs√µes geradas")

    # Logistic Regression
    y_pred_lr_proba = lr_model.predict_proba(X_test)[:, 1]
    print("‚úÖ Logistic Regression - previs√µes geradas")

    # KNN
    y_pred_knn_proba = knn_model.predict_proba(X_test)[:, 1]
    print("‚úÖ KNN - previs√µes geradas")

    # 2. Calcular AUC-ROC
    print("\nüéØ Calculando AUC-ROC...")
    from sklearn.metrics import roc_curve, roc_auc_score, auc
    import matplotlib.pyplot as plt
    import numpy as np

    # Curvas ROC
    fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test, y_pred_rf_proba)
    fpr_lr, tpr_lr, thresholds_lr = roc_curve(y_test, y_pred_lr_proba)
    fpr_knn, tpr_knn, thresholds_knn = roc_curve(y_test, y_pred_knn_proba)

    # AUC scores
    auc_rf = roc_auc_score(y_test, y_pred_rf_proba)
    auc_lr = roc_auc_score(y_test, y_pred_lr_proba)
    auc_knn = roc_auc_score(y_test, y_pred_knn_proba)

    print(f"üìà Random Forest AUC: {auc_rf:.4f}")
    print(f"üìà Logistic Regression AUC: {auc_lr:.4f}")
    print(f"üìà KNN AUC: {auc_knn:.4f}")

    # 3. Plotar curvas ROC comparativas
    print("\nüìà Plotando curvas ROC...")

    plt.figure(figsize=(12, 10))

    # Plotar cada curva ROC
    plt.plot(fpr_rf, tpr_rf, color='blue', lw=2.5,
             label=f'Random Forest (AUC = {auc_rf:.4f})', alpha=0.8)
    plt.plot(fpr_lr, tpr_lr, color='red', lw=2.5,
             label=f'Logistic Regression (AUC = {auc_lr:.4f})', alpha=0.8)
    plt.plot(fpr_knn, tpr_knn, color='green', lw=2.5,
             label=f'KNN (AUC = {auc_knn:.4f})', alpha=0.8)

    # Linha de refer√™ncia (classificador aleat√≥rio)
    plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--',
             label='Classificador Aleat√≥rio (AUC = 0.5)', alpha=0.6)

    # Configura√ß√µes do gr√°fico
    plt.xlim([-0.01, 1.01])
    plt.ylim([-0.01, 1.01])
    plt.xlabel('Taxa de Falsos Positivos (False Positive Rate)', fontsize=12, fontweight='bold')
    plt.ylabel('Taxa de Verdadeiros Positivos (True Positive Rate)', fontsize=12, fontweight='bold')
    plt.title('CURVAS ROC - DETEC√á√ÉO DE INTRUS√ïES EM REDE\nCompara√ß√£o dos Modelos',
              fontsize=14, fontweight='bold', pad=20)
    plt.legend(loc="lower right", fontsize=11)
    plt.grid(True, alpha=0.3)

    # Adicionar informa√ß√µes
    plt.text(0.6, 0.25,
             f'''üéØ INTERPRETA√á√ÉO:
    ‚Ä¢ AUC > 0.9: Excelente discriminador
    ‚Ä¢ AUC > 0.8: Bom discriminador
    ‚Ä¢ AUC = 0.5: Classificador aleat√≥rio

    üìä RESULTADOS:
    Random Forest: {auc_rf:.4f}
    Logistic Regression: {auc_lr:.4f}
    KNN: {auc_knn:.4f}''',
             bbox=dict(boxstyle="round,pad=0.5", facecolor="lightblue", alpha=0.8),
             fontsize=10, fontweight='bold')

    plt.tight_layout()
    plt.show()

    # 4. Encontrar pontos √≥timos
    print("\nüîé Encontrando thresholds √≥timos...")

    def find_optimal_threshold(fpr, tpr, thresholds):
        """Encontra o threshold √≥timo (mais pr√≥ximo do canto superior esquerdo)"""
        distances = np.sqrt(fpr**2 + (1-tpr)**2)
        optimal_idx = np.argmin(distances)
        return thresholds[optimal_idx], fpr[optimal_idx], tpr[optimal_idx]

    optimal_threshold_rf, fpr_opt_rf, tpr_opt_rf = find_optimal_threshold(fpr_rf, tpr_rf, thresholds_rf)
    optimal_threshold_lr, fpr_opt_lr, tpr_opt_lr = find_optimal_threshold(fpr_lr, tpr_lr, thresholds_lr)
    optimal_threshold_knn, fpr_opt_knn, tpr_opt_knn = find_optimal_threshold(fpr_knn, tpr_knn, thresholds_knn)

    print(f"üìè Threshold √≥timo Random Forest: {optimal_threshold_rf:.4f}")
    print(f"üìè Threshold √≥timo Logistic Regression: {optimal_threshold_lr:.4f}")
    print(f"üìè Threshold √≥timo KNN: {optimal_threshold_knn:.4f}")

    # 5. Plot detalhado com pontos √≥timos
    print("\nüé® Plotando vers√£o detalhada com pontos √≥timos...")

    plt.figure(figsize=(13, 10))

    # Curvas ROC
    plt.plot(fpr_rf, tpr_rf, color='blue', lw=2.5,
             label=f'Random Forest (AUC = {auc_rf:.4f})', alpha=0.8)
    plt.plot(fpr_lr, tpr_lr, color='red', lw=2.5,
             label=f'Logistic Regression (AUC = {auc_lr:.4f})', alpha=0.8)
    plt.plot(fpr_knn, tpr_knn, color='green', lw=2.5,
             label=f'KNN (AUC = {auc_knn:.4f})', alpha=0.8)

    # Linha de refer√™ncia
    plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--',
             label='Classificador Aleat√≥rio', alpha=0.6)

    # Pontos √≥timos
    plt.scatter(fpr_opt_rf, tpr_opt_rf, color='blue', s=100, marker='o',
                label=f'RF √ìtimo (FPR={fpr_opt_rf:.3f}, TPR={tpr_opt_rf:.3f})', edgecolors='black')
    plt.scatter(fpr_opt_lr, tpr_opt_lr, color='red', s=100, marker='s',
                label=f'LR √ìtimo (FPR={fpr_opt_lr:.3f}, TPR={tpr_opt_lr:.3f})', edgecolors='black')
    plt.scatter(fpr_opt_knn, tpr_opt_knn, color='green', s=100, marker='^',
                label=f'KNN √ìtimo (FPR={fpr_opt_knn:.3f}, TPR={tpr_opt_knn:.3f})', edgecolors='black')

    # Configura√ß√µes
    plt.xlim([-0.01, 1.01])
    plt.ylim([-0.01, 1.01])
    plt.xlabel('Taxa de Falsos Positivos (False Positive Rate)', fontsize=12, fontweight='bold')
    plt.ylabel('Taxa de Verdadeiros Positivos (True Positive Rate)', fontsize=12, fontweight='bold')
    plt.title('CURVAS ROC COM PONTOS √ìTIMOS\nDetec√ß√£o de Intrus√µes em Rede - UNSW-NB15',
              fontsize=14, fontweight='bold', pad=20)
    plt.legend(loc="lower right", fontsize=10)
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # 6. Tabela resumo
    print("\nüìã RESUMO DAS M√âTRICAS AUC-ROC:")

    from tabulate import tabulate

    summary_data = [
        ["Random Forest", f"{auc_rf:.4f}", f"{optimal_threshold_rf:.4f}", f"{fpr_opt_rf:.3f}", f"{tpr_opt_rf:.3f}"],
        ["Logistic Regression", f"{auc_lr:.4f}", f"{optimal_threshold_lr:.4f}", f"{fpr_opt_lr:.3f}", f"{tpr_opt_lr:.3f}"],
        ["KNN", f"{auc_knn:.4f}", f"{optimal_threshold_knn:.4f}", f"{fpr_opt_knn:.3f}", f"{tpr_opt_knn:.3f}"]
    ]

    headers = ["Modelo", "AUC-ROC", "Threshold √ìtimo", "FPR √ìtimo", "TPR √ìtimo"]
    print(tabulate(summary_data, headers=headers, tablefmt="grid", stralign="center"))

    # 7. Identificar melhor modelo
    best_auc = max(auc_rf, auc_lr, auc_knn)
    if best_auc == auc_rf:
        best_model_name = "Random Forest"
    elif best_auc == auc_lr:
        best_model_name = "Logistic Regression"
    else:
        best_model_name = "KNN"

    print(f"\nüèÜ MELHOR MODELO: {best_model_name} (AUC = {best_auc:.4f})")

    # 8. Salvar gr√°fico
    print("\nüíæ Salvando gr√°fico...")
    plt.figure(figsize=(12, 10))
    plt.plot(fpr_rf, tpr_rf, color='blue', lw=2.5, label=f'Random Forest (AUC = {auc_rf:.4f})')
    plt.plot(fpr_lr, tpr_lr, color='red', lw=2.5, label=f'Logistic Regression (AUC = {auc_lr:.4f})')
    plt.plot(fpr_knn, tpr_knn, color='green', lw=2.5, label=f'KNN (AUC = {auc_knn:.4f})')
    plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--', label='Classificador Aleat√≥rio')
    plt.xlabel('Taxa de Falsos Positivos')
    plt.ylabel('Taxa de Verdadeiros Positivos')
    plt.title('Curvas ROC - Detec√ß√£o de Intrus√µes em Rede')
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.3)
    plt.savefig('auc_roc_curves.png', dpi=300, bbox_inches='tight')
    plt.close()

    print("‚úÖ Gr√°fico salvo como 'auc_roc_curves.png'")
    print("\nüéâ PASSO 11 CONCLU√çDO! An√°lise AUC-ROC finalizada com sucesso!")

except Exception as e:
    print(f"‚ùå ERRO: {e}")
    print("\nüîç Informa√ß√µes para debug:")
    print(f"X_test type: {type(X_test)}")
    print(f"X_test shape: {X_test.shape}")
    print(f"y_test type: {type(y_test)}")
    print(f"y_test shape: {y_test.shape}")

    # Verificar se os modelos t√™m predict_proba
    print(f"\nüîß Capacidades dos modelos:")
    for name, model in [('RF', rf_model), ('LR', lr_model), ('KNN', knn_model)]:
        print(f"{name} - predict_proba: {hasattr(model, 'predict_proba')}")
```
